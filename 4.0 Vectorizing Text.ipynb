### 4.1 Bag of Words Model- BoW
BoW converts text into a numerical format by counting the frequency of each word in each doc, without considering word order or context.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

CountVectorizer transforms a collection of text documents into a token of matrix counts. 

```python
data = [' Most shark attacks occur about 10 feet from the beach since that is where the people are',
        'the efficiency with which he paired the socks in the drawer was quite admirable',
        'carol drank the blood as if she were a vampire',
        'giving directions that the mountains are to the west only works when you can see them',
        'the sign said there was road work ahead so he decided to speed up',
        'the gruff old man sat in the back of the bait shop grumbling to himself as he scooped out a handful of worms']
```

#### We want to initialize out CountVectorizer class

```python
count_vec = CountVectorizer()
```

#### We now apply it to our text data
The count_vec_fit holds our bag of words representation

```python
count_vec_fit = count_vec.fit_transform(data)
```

#### We want to visualize our representation as a table

```python
bag_of_words = pd.DataFrame(count_vec_fit.toarray(), columns = count_vec.get_feature_names_out())
bag_of_words
```

### TF-IDF (Term Frequency-Inverse Document Frequency)
It measures how important a word is within a collection of texts. It measures word frequency in a single doc and how common it is across all documents.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

#### We want to initialize out TfidVectorizer class

```python
tf_vec = TfidfVectorizer()
```

#### We now apply it to our text data

```python
tf_vec_fit = tf_vec.fit_transform(data)
```

```python
tfidf_bag = pd.DataFrame(tf_vec_fit.toarray(), columns = tf_vec.get_feature_names_out())
tfidf_bag
```
