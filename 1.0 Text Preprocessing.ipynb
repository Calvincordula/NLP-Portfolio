#### 1.1 Removing stop words
Stop words include articles, prepositions, conjunctions, and pronouns, etc. These words are typically filtered out to reduce noise and focus on more meaningful terms.
The primary reason for removing stop words is to improve efficiency and accuracy in tasks like search, topic modeling, or sentiment analysis.

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

##### Extract english stopwords from the nltk library

```python
en_stopwords = stopwords.words('english')  
```

```python
en_stopwords
```

```python
sentence = "How far are you willing to go and what can you offer in exchange for that which you desire?"
```

```python
sentence_no_stopword = ' '.join([word for word in sentence.split() if word not in en_stopwords])
```

```python
sentence_no_stopword
```

```python
customer_reviews = ['sam was a great help to me in the store', 
 'the cashier was very rude to me, I think her name was eleanor', 
 'amazing work from sadeen!', 
 'sarah was able to help me find the items i needed quickly', 
 'lucy is such a great addition to the team', 
 'great service from sara she found me what i wanted']
```

```python
sarahs_reviews = []
```

```python
pattern_to_find = r'sarah?'
```

```python
for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        sarahs_reviews.append(string)
```

```python
sarahs_reviews
```

#### Now we want to find reviews that end with 'a'

```python
a_reviews = []
pattern_to_find = r'^a'

for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        a_reviews.append(string)
```

```python
 a_reviews
```

#### How do we find one item or another

```python
needwant_reviews = []
pattern_to_find = r'(need|want)ed'

for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        needwant_reviews.append(string)
```

```python
needwant_reviews
```

#### Removing punctuation 
[] ->set of characters to be matched, ^ ->"not", \w ->word character, \s ->whitespaces. It means find anything that's not a word character and not a space

```python
no_punct_reviews = []
pattern_to_find =r'[^\w\s]'
```

```python
for string in customer_reviews:
    no_punct_string = re.sub(pattern_to_find, '', string)
    no_punct_reviews.append(no_punct_string)
```

```python
no_punct_reviews
```

#### 1.4 Stemming
Reducing words to their base form. Eg. Converting-> convert, stemming-> stem, etc. 
Using fewer unique words makes the data cleaner thereby reducing complexity

```python
from nltk.stem import PorterStemmer
```

PorterStemmer is a classic rule-base algorithm that reduces English words to a simpler base

```python
ps = PorterStemmer()
```

```python
connected_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects', 'prevention']
```

```python
learn_tokens = ['learn', 'learned', 'learning', 'learns', 'learners']
```

```python
like_tokens = ['likes', 'better', 'worse']
```

```python
for x in connected_tokens:
    print(x, ": ", ps.stem(x))
```

#### 1.5 Lemmatization 
Reduces a word to a meaningful base form while preserving its meaning.

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for x in connected_tokens:
    print(x, ": ", lemmatizer.lemmatize(x))
```

```python
for x in learn_tokens:
    print(x, ": ", lemmatizer.lemmatize(x))
```

```python
for x in like_tokens:
    print(x, ": ", lemmatizer.lemmatize(x))
```

Stemming is a simpler and faster process that removes suffixes to reduce words to their root forms without considering the word's context or ensuring a grammatically correct form. 
On the other hand, lemmatization is more sophisticated, using a dictionary to ensure that words are reduced to meaningful base forms considering the context. This makes lemmatization slower but more accurate than stemming.

#### 1.6 N-grams
N-gram helps analyse not only individual words but also the relationship between neighboring words. 
N-gram = a sequence of N tokens. The value of N tells how words are grouped. When N = 1, we have single words called unigram, N = 2: bigrams, N = 3: trigrams. Eg. "I love NLP" = trigram, 'I love', 'love NLP' = bigram and 'I', 'love', 'NLP' = unigram

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significant', 'advancements', 'in', 'natural', 'language', 'processing', 'computer', 'vision', 'and', 'other', 'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated', 'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought', 'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of', 'deep', 'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful', 'capable', 'of', 'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from', 'it', 'in', 'ways', 'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai', 'is', 'increasingly', 'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from', 'healthcare', 'to', 'finance', 'to', 'transportation', 'and', 'its', 'impact', 'is', 'only', 'set', 'to', 'grow', 'in', 'the', 'years', 'to', 'come']
tokens
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())
unigrams
```

```python
unigrams[:10].sort_values().plot.barh(color='lightsalmon', width=.9, figsize=(12,8))
plt.title('10 most occuring unigrams')
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
bigrams
```

```python
bigrams[:10].sort_values().plot.barh(color='orange', width=.9, figsize=(12,8))
plt.title('10 most occuring bigrams')
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())
trigrams
```
